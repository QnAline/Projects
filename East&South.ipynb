# Data cleaning

## Missing values

# 2023

```python
print(df_2023.isnull().sum())
```

```python
threshold = 0.5 # drop columns with more than 50% missing values
df_2023.dropna(axis=1, thresh=int(threshold * len(df_2023)), inplace=True)
print(df_2023.isnull().sum())
```

```python
df_2023.fillna(method='bfill', inplace=True)
print(df_2023.isnull().sum())
```

```python
# Fill missing values with the median
median_value = df_2023['COMP_L1'].median()
df_2023['COMP_L1'].fillna(median_value, inplace=True)
print(df_2023.isnull().sum())
```

# 2024

```python
print(df_2024.isnull().sum())
```

```python
threshold = 0.5 # drop columns with more than 50% missing values
df_2024.dropna(axis=1, thresh=int(threshold * len(df_2024)), inplace=True)
print(df_2024.isnull().sum())
```

```python
df_2024.fillna(method='bfill', inplace=True)
print(df_2024.isnull().sum())
```

```python
# Fill missing values with the median
median_value = df_2024['OPVBirth'].median()
df_2024['OPVBirth'].fillna(median_value, inplace=True)
median_value = df_2024['TT1_PW'].median()
df_2024['TT1_PW'].fillna(median_value, inplace=True)
print(df_2024.isnull().sum())
```

# Latitude and Longitude

```python
from geopy.geocoders import ArcGIS
from geopy.exc import GeocoderTimedOut, GeocoderServiceError
import pandas as pd
from tqdm import tqdm
import time

geolocator = ArcGIS()

def get_coordinates(district_name, max_retries=3):
    for attempt in range(max_retries):
        try:
            location = geolocator.geocode(district_name, exactly_one=True)
            if location:
                return location.latitude, location.longitude
            return None, None
        except (GeocoderTimedOut, GeocoderServiceError) as e:
            if attempt == max_retries - 1:
                print(f"Error fetching coordinates for {district_name}: {e}")
                return None, None
            time.sleep(1)

unique_districts = df_2023['District_Name'].unique()

coordinates = {}
for district in tqdm(unique_districts, desc="Fetching coordinates"):
    lat, lon = get_coordinates(district)
    coordinates[district] = (lat, lon)
    time.sleep(1)

coord_df = pd.DataFrame.from_dict(coordinates, orient='index', columns=['Latitude', 'Longitude'])
coord_df.index.name = 'District_Name'
coord_df = coord_df.reset_index()

result_df = df_2023.merge(coord_df, on='District_Name', how='left')

result_df = result_df.loc[:, ~result_df.columns.duplicated()]
result_df.to_excel('Lat_long_df_2023.xlsx', index=False)

print(result_df.head())
```
